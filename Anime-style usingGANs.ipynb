```python
!git clone https://github.com/bryandlee/animegan2-pytorch.git
%cd animegan2-pytorch
!pip install -r requirements.txt

!gdown https://drive.google.com/uc?id=1styEFeMY13hP7VZz4CVP6ojF8Ydv4jYh -O weights/celeba_distill.pt

from google.colab import files
uploaded = files.upload()  # Upload your image when prompted

import torch
from torchvision import transforms
from PIL import Image
from model import Generator

device = 'cuda' if torch.cuda.is_available() else 'cpu'

model = Generator().to(device)
model.load_state_dict(torch.load('weights/celeba_distill.pt'))
model.eval()

input_image_path = next(iter(uploaded))

img = Image.open(input_image_path).convert("RGB")
transform = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.ToTensor(),
])
tensor = transform(img).unsqueeze(0).to(device)

with torch.no_grad():
    output = model(tensor).cpu()

out_img = transforms.ToPILImage()(output.squeeze())
out_img.save("anime_output.png")
out_img.show()
```
